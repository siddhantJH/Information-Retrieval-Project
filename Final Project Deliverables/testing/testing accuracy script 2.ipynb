### Rouge Score

```python
from rouge import Rouge
def rougeSimilarity(a, b):
    rouge = Rouge()
    getScore = rouge.get_scores(a, b)
    r1 = getScore[0]['rouge-1']['f']
    r2 = getScore[0]['rouge-2']['f']
    r3 = getScore[0]['rouge-l']['f']
    avg = (r1 + r2 + r3) / 3.0
    return avg
```

```python
data2['New Rouge Score'] = data2.apply(lambda x: rougeSimilarity(x['Original Answer'], x['new']), axis=1)
data2['Old Rouge Score'] = data2.apply(lambda x: rougeSimilarity(x['Original Answer'], x['old']), axis=1)
```

```python
sum_new_rouge = data2["New Rouge Score"].sum()
count_rouge = data2.shape[0]
acc_new_rouge = (sum_new_rouge/count_rouge)*100
sum_old_rouge = data2["Old Rouge Score"].sum()
acc_old_rouge = (sum_old_rouge/count_rouge)*100
```

```python
acc_new_rouge
```

### Jaccard Similarity

```python
import nltk
def jaccardSimilarity(str1, str2):
    i = len(str1.intersection(str2))
    u = len(str1.union(str2))
    return i / u if u != 0 else 0
```

```python
data2["New Jaccard Similarity"] = data2.apply(lambda x: jaccardSimilarity(set(nltk.word_tokenize(x['Original Answer'].lower())), set(nltk.word_tokenize(x['new'].lower()))), axis=1)
data2["Old Jaccard Similarity"] = data2.apply(lambda x: jaccardSimilarity(set(nltk.word_tokenize(x['Original Answer'].lower())), set(nltk.word_tokenize(x['old'].lower()))), axis=1)
```

```python
sum_new_jaccard = data2["New Jaccard Similarity"].sum()
count_jaccard = data2.shape[0]
acc_new_jaccard = (sum_new_jaccard/count_jaccard)*100
sum_old_jaccard = data2["Old Jaccard Similarity"].sum()
acc_old_jaccard = (sum_old_jaccard/count_jaccard)*100
```

```python
acc_new_jaccard
```

### BLEU Score

```python
from collections import Counter
import math

def calculateVal(c, r, n):
    res1 = [tuple(c[i:i+n]) for i in range(len(c)-n+1)]
    res2 = [tuple(r[i:i+n]) for i in range(len(r)-n+1)]
    if len(res1) == 0:
        return 0
    else:
        i = sum((Counter(res1) & Counter(res2)).values())
        return i / len(res1)

def bleuScore(c, r, maxi=4, weights=[0.25, 0.25, 0.25, 0.25]):
    p = [0] * maxi
    for i in range(maxi):
        p[i] = calculateVal(c, r, i+1)
    
    if min(p) == 0:
        return 0
    
    num = min(1, len(c) / len(r))
    
    score = num * math.exp(sum(weights[i] * math.log(j) for i, j in enumerate(p)))
    
    return score
```

```python
data2["New BLEU Score"] = data2.apply(lambda x: bleuScore(x['Original Answer'].split(), x['new'].split()), axis=1)
data2["Old BLEU Score"] = data2.apply(lambda x: bleuScore(x['Original Answer'].split(), x['old'].split()), axis=1)
```

```python
sum_new_bleu = data2["New BLEU Score"].sum()
count_bleu = data2.shape[0]
acc_new_bleu = (sum_new_bleu/count_bleu)*100
sum_old_bleu = data2["Old BLEU Score"].sum()
acc_old_bleu = (sum_old_bleu/count_bleu)*100
```

### Cosine Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def cosineSimilarity(str1, str2):
    vectorizer = CountVectorizer()

    X = vectorizer.fit_transform([str1, str2])

    similarity = cosine_similarity(X[0], X[1])[0][0]
    return similarity
```

```python
data2['New Cosine Score'] = data2.apply(lambda x: cosineSimilarity(x['Original Answer'], x['new']), axis=1)
data2['Old Cosine Score'] = data2.apply(lambda x: cosineSimilarity(x['Original Answer'], x['old']), axis=1)
```

```python
sum_new_Cosine_Similarity = data2["New Cosine Score"].sum()
count_Cosine_Similarity = data2.shape[0]
acc_new_Cosine_Similarity = (sum_new_Cosine_Similarity/count_Cosine_Similarity)*100
sum_old_Cosine_Similarity = data2["Old Cosine Score"].sum()
acc_old_Cosine_Similarity = (sum_old_Cosine_Similarity/count_Cosine_Similarity)*100
```
