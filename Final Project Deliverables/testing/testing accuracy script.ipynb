```python
from rouge import Rouge
def findSimilarity(a, b):
    rouge = Rouge()
    getScore = rouge.get_scores(a, b)
    r1 = getScore[0]['rouge-1']['f']
    r2 = getScore[0]['rouge-2']['f']
    r3 = getScore[0]['rouge-l']['f']
    avg = (r1 + r2 + r3) / 3.0
    return avg
```

```python
data1['New Rouge Score'] = data1.apply(lambda x: findSimilarity(x['Original'], x['New']), axis=1)
data1['Old Rouge Score'] = data1.apply(lambda x: findSimilarity(x['Original'], x['Old']), axis=1)
```

```python
data2['New Rouge Score'] = data2.apply(lambda x: findSimilarity(x['Original Answer'], x['new']), axis=1)
data2['Old Rouge Score'] = data2.apply(lambda x: findSimilarity(x['Original Answer'], x['old']), axis=1)
```
